# Plotting

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn, rand
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
arr1 = np.random.randn(1000)
arr2 = np.random.randn(1000)
```

```python
plt.hist(arr1, density = True, color = "crimson", alpha=0.5, bins = 200)
plt.hist(arr2, density = True, color = "olive", alpha=0.5, bins = 200)
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
sbn.jointplot(data1,data2, kind="hex")
```
