# Aggregate Functions

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
data = pd.read_csv("Wine_quality_dataset.csv")
```

```python
data.head()
```

### Max Value

```python
data["alcohol"].max()
```

### Creating Groupby

```python
wine1= data.groupby("quality")
```

```python
wine1
```

```python
for group in wine1:
    for item in group:
        if type(item)==int:
            print("Quality: " + str(item))
        else:
            print(item[["fixed acidity", "volatile acidity", "citric acid"]])
            
# This prints the relevant data using the output of the groupby function. 
# This function works because of the structure of the groupby method
                
```

```python
data.describe()
```

```python
wine1.describe()
```

## Creating an Aggregate Function

```python
def max_to_min(arr):
    return arr.max()-arr.min()
```

```python
wine1.agg(max_to_min)
```

## Adding a new calculated column

```python
data["free/total sulfur dioxide"] = data["free sulfur dioxide"]/data["total sulfur dioxide"]
```
