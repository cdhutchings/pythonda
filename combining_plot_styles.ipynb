# Combining Plot Styles

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sbn
import matplotlib as mpl
from numpy.random import randn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
number = randn(100)
```

```python
sbn.distplot(number,bins=20)
```

```python
sbn.distplot(number, rug=True,hist=False)
```

```python
sbn.distplot(number,bins=15)
```

```python
dataset = pd.read_csv("weather_data.csv")
sbn.lmplot("Temperature", "WindSpeed", number, order=4,
          scatter_kws = {"marker": "o"})
```

```python
tips = sbn.load_dataset("tips")
tips.tail()
```

```python
# Heat maps
```

```python
flight_dframe = sbn.load_dataset("flights").pivot("year", "month","passengers")
flight_dframe.head()
```

```python
sbn.heatmap(flight_dframe)
```

```python
sbn.heatmap(flight_dframe,annot=True, fmt="d")
```

```python
sbn.heatmap(flight_dframe,center=flight_dframe.loc["July",1960])
```

```python
sbn.clustermap(flight_dframe,annot=True,fmt="d")
```

```python
sbn.clustermap(flight_dframe,col_cluster=False)
```

```python
sbn.clustermap(flight_dframe,standard_scale=1)
```

```python
sbn.clustermap(flight_dframe,standard_scale=0)
```
